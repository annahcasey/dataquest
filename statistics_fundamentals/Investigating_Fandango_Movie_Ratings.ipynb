```python
fd38 = five38[['FILM', 'Fandango_Stars', 'Fandango_Ratingvalue', 'Fandango_votes', 'Fandango_Difference']]
fddq = dq[['movie', 'year', 'fandango']]
```

```python
fddq = fddq[fddq['year']==2016]
```

```python
fd38 = fd38[fd38['FILM'].str.contains(r'(2015)')]
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(16,9))
ax = plt.gca()
sns.kdeplot(fd38['Fandango_Stars'], bw=.2)
sns.kdeplot(fddq['fandango'], bw=.2)
plt.title('Distribution of Fandango Film Ratings', size=32)
ax.set_xlabel('Star Rating')
ax.set_ylabel('Relative Frequency')
plt.legend(['2015','2016'])
ax.set_xticks([i/2 for i in range(0,10)])
ax.set_xlim(0,5)
```

```python
fd38['Fandango_Stars'].value_counts(normalize=True).sort_index()*100
```

```python
fddq['fandango'].value_counts(normalize=True).sort_index()*100
```

```python
mean38 = fd38['Fandango_Stars'].mean()
```

```python
median38 = fd38['Fandango_Stars'].median()
```

```python
mode38 = fd38['Fandango_Stars'].mode()[0]
```

```python
meandq = fddq['fandango'].mean()
```

```python
mediandq = fddq['fandango'].median()
```

```python
modedq = fddq['fandango'].mode()[0]
```

```python
summary_stats = pd.DataFrame([mean38, meandq, median38, mediandq, mode38, modedq])
```

```python
summary_stats['year']=[2015,2016,2015,2016,2015,2016]
```

```python
summary_stats['stat']=['mean', 'mean', 'median', 'median', 'mode', 'mode']
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(16,9))
ax = plt.gca()
sns.barplot(x='stat', y=0, hue='year',data=summary_stats)
ax.set_ylabel('Star Rating')
ax.set_xlabel('Summary Stat')
ax.set_yticks([i/2 for i in range(0,10)])
ax.set_ylim(0,5)
plt.title('Fandango Star Rating Summary Statistics for 2015 vs. 2016', size=32)
```
