```python
%%sql
SELECT a1.title AS album,
       g.name AS genre
  FROM track AS t
       LEFT JOIN album AS a1
       ON a1.album_id = t.album_id
       LEFT JOIN genre AS g
       ON g.genre_id = t.genre_id
 GROUP BY genre, album
 ORDER BY 'value_occurrence' DESC
```

We've been given the opportunity to buy four records from a new label, though we only have the funds to get three of them. Since this label's market is the USA, and we only know the genres of the albums, the task is to find the most popular genres in the USA by number of tracks purchased. However, a more relevant question might be how much revenue each genre has generated in the USA. I shall try to address both. 

```python
%%sql
SELECT g.name AS genre,
       SUM(il.quantity) AS number_of_purchases,
       100* (CAST(SUM(il.quantity) AS FLOAT) / 
       (SELECT CAST(SUM(il.quantity) AS FLOAT)
          FROM invoice_line AS il
               LEFT JOIN invoice AS i
               ON i.invoice_id = il.invoice_id
               LEFT JOIN customer AS c
               ON c.customer_id = i.customer_id
         WHERE c.country = 'USA')) AS percentage_of_purchases
  FROM invoice_line AS il
       LEFT JOIN invoice AS i
       ON i.invoice_id = il.invoice_id
       LEFT JOIN customer AS c
       ON c.customer_id = i.customer_id
       LEFT JOIN track AS t
       ON t.track_id = il.track_id
       LEFT JOIN genre AS g
       ON g.genre_id = t.genre_id
 WHERE c.country = 'USA'
 GROUP BY genre
 ORDER BY number_of_purchases DESC;
       
```

```python
%%sql
SELECT g.name AS genre,
       SUM(i.total) AS total_revenue
  FROM invoice_line AS il
       LEFT JOIN invoice AS i
       ON i.invoice_id = il.invoice_id
       LEFT JOIN customer AS c
       ON c.customer_id = i.customer_id
       LEFT JOIN track AS t
       ON t.track_id = il.track_id
       LEFT JOIN genre AS g
       ON g.genre_id = t.genre_id
 WHERE c.country = 'USA'
 GROUP BY genre
 ORDER BY total_revenue DESC;
```

Based solely on number of purchases, I would suggest dropping the hip-hop album from our list, since the other three genres (pop, blues, and punk) seem to be more popular. However, if we were interested in which genres generate the most revenue for us in the USA, it would make more sense to drop the pop album. Despite the narrow margin of popularity pop has over hip-hop, hip-hop has historically generated more revenue for us. 

```python
%%sql
SELECT e1.first_name||' '||e1.last_name AS employee_name,
       e2.first_name||' '||e2.last_name AS supervisor_name,
       e1.hire_date,
       e1.title,
       SUM(i.total) AS attributable_sales_revenue
  FROM invoice AS i
       LEFT JOIN customer AS c
       ON c.customer_id = i.customer_id
       LEFT JOIN employee AS e1
       ON e1.employee_id = c.support_rep_id
       LEFT JOIN employee AS e2
       ON e2.employee_id = e1.reports_to
 GROUP BY e1.employee_id;
```

```python
%%sql
  WITH uncollapsed_other AS
       (SELECT CASE
               WHEN COUNT(DISTINCT c.customer_id) = 1 THEN 'Other' 
               ELSE c.country 
               END AS country,
               COUNT(DISTINCT c.customer_id) AS number_of_customers,
               SUM(i.total) AS total_sales_revenue,
               SUM(i.total) / COUNT(DISTINCT c.customer_id) AS customer_average_revenue,
               AVG(i.total) AS order_average_revenue
          FROM customer as c
               LEFT JOIN invoice AS i
               ON i.customer_id = c.customer_id
         GROUP BY country)
SELECT country,
       SUM(number_of_customers) AS customer_tally,
       ROUND(SUM(total_sales_revenue), 2) AS revenue_total,
       ROUND(AVG(customer_average_revenue), 2) AS customer_revenue_average,
       ROUND(AVG(order_average_revenue), 2) AS order_revenue_average
  FROM uncollapsed_other
 GROUP BY country
 ORDER BY CASE
          WHEN country = 'Other' then 0
          ELSE revenue_total 
          END DESC;
```

```python
%%sql        
  WITH first_track AS
       (SELECT MIN(il.track_id) AS first_track_id,
               il.invoice_id,
               t.album_id
          FROM invoice_line AS il
               LEFT JOIN track AS t
               ON t.track_id = il.track_id
         GROUP BY il.invoice_id),
       track_list AS
       (SELECT ft.invoice_id,
               t.track_id
          FROM first_track AS ft
               LEFT JOIN track AS t
               ON t.album_id = ft.album_id),
       invoice_list AS 
       (SELECT invoice_id,
               track_id
          FROM invoice_line),
       tl_except_il AS
       (SELECT *
          FROM track_list

        EXCEPT

        SELECT *
          FROM invoice_list),
       il_except_tl AS
       (SELECT *
          FROM invoice_list
       
        EXCEPT
       
        SELECT *
          FROM track_list)
SELECT COUNT (DISTINCT invoice_id) AS Nonalbum_invoice_tally,
       (SELECT COUNT (DISTINCT invoice_id)
          FROM invoice_line) - COUNT (DISTINCT invoice_id) AS Album_invoice_tally,
       (SELECT COUNT (DISTINCT invoice_id)
          FROM invoice_line) AS Total_invoice_tally,
       (100 *
       ((SELECT COUNT (DISTINCT invoice_id)
          FROM invoice_line) - COUNT (DISTINCT invoice_id))/
       ((SELECT COUNT (DISTINCT invoice_id)
          FROM invoice_line)))||'%' AS Nonalbum_invoice_percentage
  FROM (SELECT *
          FROM tl_except_il

         UNION

        SELECT *
          FROM il_except_tl);
```
