```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#imports in necessary libraries
%matplotlib inline
#sets up matplotlib to display inline
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
#reads in csv
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
#creates a couple of colors from their RGB values
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
#list of bachelor degree majors
fig = plt.figure(figsize=(10, 20))
#creates a 18x3 figure
for sp in range(-2,6):
    ax = fig.add_subplot(6,3,sp+3)
    #creates a subplot in figure for each category in stem_cats
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=4)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=4)
    #plots the Women vs. Men lines for each category
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    #removes spines
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #sets x and y limits
    ax.set_title(stem_cats[sp])
    #gives each subplot a title
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    #removes ticks
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')

    elif sp == 2:
        ax.text(2002, 82, 'Women')
        ax.text(2005, 15, 'Men')
    
    elif sp == 4:
        ax.text(2005, 62, 'Men')
        ax.text(2002, 32, 'Women')
    #labels the lines in the first subplot for each row
plt.show()
#displays the figure
```
