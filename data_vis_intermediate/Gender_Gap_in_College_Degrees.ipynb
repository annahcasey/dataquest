```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#imports in necessary libraries
%matplotlib inline
#sets up matplotlib to display inline
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
#reads in csv
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
#creates a couple of colors from their RGB values
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
#lists of bachelor degree majors by category
fig = plt.figure(figsize=(20,20))
#creates a 18x3 figure
for sp in range(0,6):
    ax = fig.add_subplot(6,3,(sp*3+1))
    #creates a subplot in figure for each category in stem_cats in the first column
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=4)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=4)
    #plots the Women vs. Men lines for each category
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    #removes spines
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #sets x and y limits
    ax.set_title(stem_cats[sp])
    #gives each subplot a title
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    #removes ticks and x-axis labels
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    #sets y-ticks to just display 0 and 100 and creates horizontal line at 50
    if sp == 0:
        ax.text(2005, 88, 'Men')
        ax.text(2003, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 60, 'Men')
        ax.text(2003, 35, 'Women')
        ax.tick_params(labelbottom=True)
    #adds labels for top and bottom plots in this column and re-assigns x-axis labels to bottom plot
for sp in range(0,5):
    ax = fig.add_subplot(6,3,(sp*3+2))
    #creates a subplot in figure for each category in lib_arts_cats in the second column
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=4)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=4)
    #plots the Women vs. Men lines for each category
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    #removes spines
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #sets x and y limits
    ax.set_title(lib_arts_cats[sp])
    #gives each subplot a title
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    #removes ticks and x-axis labels
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    #sets y-ticks to just display 0 and 100 and creates horizontal line at 50
    if sp == 0:
        ax.text(2005, 22, 'Men')
        ax.text(2003, 75, 'Women')
    elif sp == 4:
        ax.tick_params(labelbottom=True)
    #adds labels for the topmost plot in this column (bottom has overlapping lines), and re-assigns x-axis labels to bottom plot
for sp in range(0,6):
    ax = fig.add_subplot(6,3,(sp*3+3))
    #creates a subplot in figure for each category in other_cats in the last column
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=4)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=4)
    #plots the Women vs. Men lines for each category
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    #removes spines
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #sets x and y limits
    ax.set_title(other_cats[sp])
    #gives each subplot a title
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    #removes ticks and x-axis labels
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    #sets y-ticks to just display 0 and 100 and creates horizontal line at 50
    if sp == 0:
        ax.text(2005, 7, 'Men')
        ax.text(2003, 90, 'Women')
    elif sp == 5:
        ax.text(2005, 61, 'Men')
        ax.text(2003, 35, 'Women')
        ax.tick_params(labelbottom=True)
    #adds labels for top and bottom plots in this column, and re-assigns labels to x-axis of bottom plot

plt.savefig("gender_degrees.png")
#displays the figure
```
